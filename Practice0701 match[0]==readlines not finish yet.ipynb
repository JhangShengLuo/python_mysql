#把同義詞建成字典
file1 = open('category.txt','r')
dic={}
counter = 0
for i in file1.readlines():
    if file1 != '\n':
        counter += 1
    if '' in i.split(','): 
        a = i.split(',')[:i.split(',').index('')] #如果有空白的話，把後面空白截掉
        for j in a[1:]:
            dic[j] = a[0] #把同義詞對到食材
    else:
        a = i.split(',') 
        for j in a[1:]:
            dic[j] = a[0]
            
print '食材tag數量： ', counter
QQQ = len(dic)-counter
print '食材（含同義字）數量： ', QQQ
