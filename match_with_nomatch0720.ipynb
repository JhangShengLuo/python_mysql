```python
import json
import MySQLdb
import pandas

try:
    db = MySQLdb.connect("10.120.26.29","root","password","food",charset='utf8')
 
    # 執行SQL statement
    cursor = db.cursor()
    cursor.execute("SELECT * FROM samefood;")
    
    # 撈取多筆資料
    results = cursor.fetchall()
    
#     把samefood的資料撈出來

    
    samefoodList=[]
     # 迴圈撈取資料
    for record in results:
        samefoodList.append(record[1])
#   去Nutrition表抓資料
    cursorN = db.cursor()
    cursorN.execute("SELECT * FROM nutrition;")
    resultsN = cursorN.fetchall()


# --------------------------------------------------
    
    
    match100=[] #吻合的建一個list
    match =[]
    nomatch100=[]   #沒有完全吻合的建一個LIST
    for recordN in resultsN:
        if recordN[2] in samefoodList:
            match.append(recordN[2])
    for recordN in resultsN:
        if recordN[2] in samefoodList:
            match100.append([recordN[2],recordN[3],recordN[4],recordN[5],recordN[6],recordN[7],recordN[8]])
    
    print len(match),'完全吻合的數量'
    
    with open('match100.txt','w') as f:
        f.write('樣品名稱,熱量成分值kcal,水分成分值g,粗蛋白成分值g,粗脂肪成分值g,總碳水化合物成分值g,膳食纖維成分值g,')
        for matchRow in match100:
#         print matchrRow
            f.write('\n')
            for ele in matchRow:
#            print ele  #把吻合的食材及熱量印出來
                f.write(format(ele).encode('utf8')+',')
    
    #把沒有吻合名單建立出來
    for element in samefoodList:
         if element not in match:
            nomatch100.append(element)
#     for element in  nomatch100:
#         print element
    print len(nomatch100),'沒完全吻合的'

    


#--------------------------------------------------
    
    
    print "---------------------------------------"
    print "---------------------------------------"
    print "---------------------------------------"
    #--------------------------------------------------------
 
    nutritionList=[] #建一個營養成分list
    
    for recordN in resultsN:
        nutritionList.append(recordN[2])
    
  
    matchSet = set() #不會重複的collections
    uniqueMatch = [] #有吻合就算的LIST
   
    
    for eleS in nomatch100:
        for eleN in nutritionList:
            if (eleS in eleN):
                matchSet.add(eleS)
#                 print eleS,eleN
    for eleML in matchSet:
        uniqueMatch.append(eleML)
        
   
    
    
    print "--------------------"
    print "--------------------"
    print "--------------------"
    
    
    print len(uniqueMatch),'有一些吻合'
    with open('uniqueMatch.txt','w') as f:
        f.write('樣品名稱,熱量成分值kcal,水分成分值g,粗蛋白成分值g,粗脂肪成分值g,總碳水化合物成分值g,膳食纖維成分值g,'+'\n')
        for ele in uniqueMatch:
            f.write(format(ele).encode('utf8')+','+'\n')
        
        
        
    totalNomatch = [] #完全沒吻合的食材LIST
#     for element in uniqueMatch:
#         print element
    for element in nomatch100:
        if element not in uniqueMatch:
            totalNomatch.append(element)
            
    print len(totalNomatch),'完全沒吻合'
    with open('totalNomatch.txt','w') as f:
        f.write('樣品名稱,熱量成分值kcal,水分成分值g,粗蛋白成分值g,粗脂肪成分值g,總碳水化合物成分值g,膳食纖維成分值g,'+'\n')
        for ele in totalNomatch:
            f.write(format(ele).encode('utf8')+','+'\n')
    
   

    # 關閉連線
    db.close()

except MySQLdb.Error as e:
    print "Error %d: %s" % (e.args[0], e.args[1])
    
    
    
    
    
    
```
